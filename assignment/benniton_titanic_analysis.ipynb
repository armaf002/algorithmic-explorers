```python
df['Age'] = df['Age'].apply(lambda x: str(x))
df['Age'] = df['Age'].str.replace('nan', '')
df
```

```python
df['Cabin'] = df['Cabin'].apply(lambda x: str(x))
df['Cabin'] = df['Cabin'].str.replace('nan', '')
df
```

```python
df['Embarked'] = df['Embarked'].apply(lambda x: str(x))
df['Embarked'] = df['Embarked'].str.replace('nan', '')
df
```

```python
df.isnull().sum()
```

```python
df['Name'] = df['Name'].str.replace(r'\(.*?\)', '', regex=True)
df
```

```python
df['Name'] = df['Name'].str.replace(r'"[^"]*"', '', regex=True)
df
```

```python
# Function to move text after the comma to the front
def name_reposition(text):
    if ',' in text:  # Check if the string contains a comma
        parts = text.split(',', 1)  # Split the string into two parts at the first comma
        return parts[1].strip() + ' ' + parts[0].strip()  # Move text after comma to the front
    return text  # Return unchanged if no comma is present

# Apply the function to the 'Name' column
df['Name'] = df['Name'].apply(name_reposition)
df
```
