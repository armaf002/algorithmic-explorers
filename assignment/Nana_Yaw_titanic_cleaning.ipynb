### Importing Necessary libraries and modules

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# Importing dataset, setting index as the passenger id and making a copy
titanic_main = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/refs/heads/master/titanic.csv",index_col=0)
titanic = titanic_main.copy()
```

### EDA

```python
# Viewing the top 10 observation
titanic.head(10)
```

```python
#viewing the bottom 10 observation
titanic.tail(10)
```

```python
#Checking datatypes of columns
titanic.info()
```

```python
#Checking the number of null values in each column
titanic.isnull().sum()
```

```python
#Removing the cabin column since it has a lot of null values
titanic = titanic.drop('Cabin',axis=1)
```

```python
titanic.head()
```

```python
titanic.Age = titanic.Age.fillna(titanic.Age.mean().round())
```

```python
# Filling the null values in the embarked column with the mode
titanic.Embarked.fillna(titanic.Embarked.mode()[0],inplace=True)
```

```python
#changing the age column to int
titanic.Age = titanic.Age.astype('int64')
```
