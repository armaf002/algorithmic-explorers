# HANDLING MISSING VALUES IN THE AFFECTED COLUMN

# Handling Missing Values in the Age column

So to make my decision on what to do with the missing value of this column, I had to check if dropping the column would affect other columns.
1. The idea was to to check the ratio of categorical columns before and after dropping the missing age
2. For the fare column i'll check its distribution before and afte dropping the missing age

```python
missingAge = titanic[titanic['Age'].isnull()]
titanicAfter = titanic[~titanic['Age'].isnull()]
```

```python
display(titanic['Sex'].value_counts())
print(f'ratio of female to male in titanic: {314/577}')

display(missingAge['Sex'].value_counts())
print(f'ratio of female to male in missing age: {53/124}')

display(titanicAfter['Sex'].value_counts())
print(f'ratio of female to male in titanic after dropping missing age: {261/453}')
```

```python
display(titanic['Survived'].value_counts())
print(f'ratio of Survival in titanic: {342/549}')

display(missingAge['Survived'].value_counts())
print(f'ratio of Survival in missing age: {52/125}')

display(titanicAfter['Survived'].value_counts())
print(f'ratio of Survival in titanic after droping missing age: {290/424}')
```

```python
# Comparing for the P class Column
display(titanic['Pclass'].value_counts())

display(missingAge['Pclass'].value_counts())

display(titanicAfter['Pclass'].value_counts())
```

```python
#distribution of fare before dropping missing age
plt.hist(titanic['Fare'], bins=10)
```

```python
#distribution of fare before dropping missing age
plt.hist(titanicAfter['Fare'], bins=10)
```

```python
# I decided to drop the rows with missing age because it has little effect on the dataset
titanic = titanic.dropna(subset=['Age'])
```

# Handling Missing Values in the Cabin Column

While checking out each column,I concluded that cabin doesn't have much effect on the whole dataset. Except that i noticed that most people who are related stay in the same cabin.
But Ticket column also shows that people who are related had the same ticket Id. So Ticket column can do the same job as Cabin column, so decided to drop the whole column

# Handling Missing Values in the Embarked Column

# Visualizing the distribution of both Categorical and Numeric Variables

```python
Numeric = titanic.select_dtypes(include = ['float64'])
Numeric['SibSp'] = titanic['SibSp']
Numeric['Parch'] = titanic['Parch']
Categorical = titanic.select_dtypes(include = ['object', 'int64']).drop(['PassengerId', 'Name', 'SibSp', 'Parch'], axis=1)
```

```python
Numeric.hist(figsize=(15, 5), bins = 10)
```

Conclusion:
1. The age column almost normally distributed. With the age of most passengers being form 18 to 30
2. This means that most of the passengers were young people
3. But also there is a significant amount of little chidren/babies on this ship. About 50 of them.
4. It seems that most of the passengers paid little to nothing. But i don't know the currency so i can't say much
5. Most of the pasengers did not come with family, that is their siblings, spouse, parents or children

```python
# Showing distribution of Categorical Variables
fig, axs = plt.subplots(2, 3, figsize = (15, 2*3))
for i, col in enumerate(Categorical.columns):
    row = i // 3
    col_idx = i% 3
    sns.countplot(data = Categorical, x = col, ax = axs[row, col_idx])
plt.tight_layout()
plt.show()
```

Conclusion:
1. More than a hundred more people survived this voyage than not
2. Most people on titanic bought the 3rd class ticket. Almost the same ammount of people bought second class and first class ticket
3. There are more male than female on this voyage
4. Most people embarked on S town
